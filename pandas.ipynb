```python
# isin()
# Find the most destructive death over batsman in the history of ipl
# Min batsman 200 balls in 16-20
```

```python
death_over = delivery[delivery['over'] > 15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
```

```python
batsman_list = all_batsman[all_batsman > 200].index.tolist()
```

```python
# Runs scored by all these 43 batsman
# Balls played by all these 43 batsman
```

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = delivery.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = delivery.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls) * 100
```

```python
sr.sort_values(ascending = False).head(10)
```
